```python
import firebase_admin
from firebase_admin import credentials, firestore
import json

# Initialize Firebase
cred = credentials.Certificate("guia-e1e64-firebase-adminsdk-yw4u3-9f3a257320.json")
#   firebase_admin.initialize_app(cred)

# Initialize Firestore
db = firestore.client()

#read the data from the json file
with open('repo_content.json') as f:
    data = json.load(f)


# Store each repository and its files
for repo in data:
    print(repo['repository_url'])
    # Add the repository as a document in the 'repositories' collection
    repo_ref = db.collection('repositories').document()  # You can specify an ID or let Firestore generate one
    
    # Store the repository URL
    repo_ref.set({
        'repository_url': repo['repository_url']
    })
    
    # Store each file as a document in a 'files' subcollection within the repository document
    for file in repo['files']:
        repo_ref.collection('files').add(file)
```

```python
from google.cloud import storage
def create_bucket_class_location(bucket_name):
    """
    Create a new bucket in the US region with the coldline storage
    class
    """

    storage_client = storage.Client()

    bucket = storage_client.bucket(bucket_name)
    bucket.storage_class = "COLDLINE"
    new_bucket = storage_client.create_bucket(bucket, location="us")

    print(
        "Created bucket {} in {} with storage class {}".format(
            new_bucket.name, new_bucket.location, new_bucket.storage_class
        )
    )
    return new_bucket
create_bucket_class_location("guia-storage")
```
